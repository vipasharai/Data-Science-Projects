```python
# What is the median salary of male and female employees
df['Salary'].value_counts()
```

```python
df.groupby('Sex')['Salary'].median()
```

```python
# What is the total absences and average engagement survey score for each department
df.groupby('Department').agg({'Absences':'sum','EngagementSurvey':'mean'})
```

```python
# What is total number of special projects and average absence for employees in each gender 
# category 
```

```python
df.groupby('Sex').agg({'Absences':'mean','SpecialProjectsCount':'sum'})
```

```python
# What is maximum salary and minimum days late in the last 30 days for employees in each position
```

```python
df.groupby('Position').agg({'Salary':'max','DaysLateLast30':'min'})
```

```python
# what is earliest and latest date of hire for employee in each manager team ?
df.groupby('ManagerName')['DateofHire'].agg(['min','max'])
```

```python
# how many terminated employees were there in each department and what is average employee satisfaction  label in team 
df.groupby('Department').agg({'Termd':'sum',"EmpSatisfaction":'mean'})
```
